***Problem Statement:***
Determine the concentration of the four greenhouse gases: CO$_{2}$, H$_{2}$0, CH$_{4}$, and N$_{2}$O in Earth's atmosphere.

***Variables and Parameters***

*We begin by defining our variables and parameters, where the concentration of each gas is defined by its mass over its volume, and its initial concentration is some constant following the notation $C_{gas\_name, 0}$. Moreover, we specify the rate of change in greenhouse gases for each source (wetlands, plants, etc.), assuming that the rate of change of each individual greenhouse gas will be equal for each source, as is reported in our assumptions and constraints.*

$$
\begin{align}
\begin{array}{c}
\textbf{Description} & \textbf{Symbol} & \textbf{Dimensions} & \textbf{Type} \\
\hline \text{Concentration of CO$_{2}$ in the atmosphere} & C_{CO_{2}} & \text{M L}^{-3} & \text{dependent variable} \\
\hline \text{Concentration of CH$_{4}$ in the atmosphere} & C_{CH_{4}} & \text{M L}^{-3} & \text{dependent variable} \\
\hline \text{Concentration of H$_{2}$O in the atmosphere} & C_{H_{2}O} & \text{M L}^{-3} & \text{dependent variable} \\
\hline \text{Concentration of N$_{2}$O in the atmosphere} & C_{N_{2}O} & \text{M L}^{-3} & \text{dependent variable} \\
\hline \text{Time} & t & \text{T} & \text{independent variable} \\
\hline \text{Rate of change of volume of greenhouse gases due to volcanic eruptions} & R_{v} & \text{M T}^{-1} & \text{parameter} \\
\hline \text{Rate of change of volume of greenhouse gases due to wetlands} & R_{w} & \text{M T}^{-1} & \text{parameter} \\
\hline \text{Rate of change of volume of greenhouse gases due to the ocean} & R_{o} & \text{M T}^{-1} & \text{parameter} \\
\hline \text{Rate of change of volume of greenhouse gases due to plant life} & R_{p} & \text{M T}^{-1} & \text{parameter} \\
\hline \text{Rate of change of volume of CH$_{4}$ breaking into CO$_{2}$ and H$_{2}$O} & R_{m} & \text{M T}^{-1} & \text{parameter} \\

\hline \text{Rate of change of volume of water vapour precipitation} & R_{pr} & \text{M T}^{-1} & \text{parameter} \\
\hline \text{Initial concentration of CO$_{2}$ in the atmosphere} & C_{CO_{2},0} & \text{M L}^{-3} & \text{parameter} \\
\hline \text{Initial concentration of CH$_{4}$ in the atmosphere} & C_{CH_{4},0} & \text{M L}^{-3} & \text{parameter} \\
\hline \text{Initial concentration of H$_{2}$O in the atmosphere} & C_{H_{2}O,0} & \text{M L}^{-3} & \text{parameter} \\
\hline \text{Initial concentration of N$_{2}$O in the atmosphere} & C_{N_{2}O,0} & \text{M L}^{-3} & \text{parameter} \\
\hline
\end{array}
\end{align}
$$

**Assumptions and Constraints**

*Next, we define our assumptions and constraints, ensuring our system takes into account all necessary factors, and describes any relevant requirements for our model to function more realistically.*

- The atmosphere is a well-mixed fluid
- The atmosphere is a fluid of constant/negligible mass and constant volume
- The rates of all processes for all chemicals is constant
- Any processes not stated in the variables and parameters that could affect the concentration of our chemicals are treated as negligible
- Each process releases all chemicals in equal amounts
- Assume the initial concentration of each gas to be equal ($C_{CO_2,0} = C_{H_2O,0} = C_{CH_4,0} = C_{N_2O,0}$)

**Constructing the model**

*Now, we start building our solution, beginning with the initial equations derived from the law of conservation of mass: $\frac{dM}{dt} = Q_{in} - Q_{out}$. We track the rate of change of mass of each individual gas separately given the different emission/absorbtion rates for each gas. Therefore, we define $Q_{in}$ to be the mass of each gas entering our atmosphere, and $Q_{out}$ to be the mass of each gas exiting our atmosphere.*

$$
\begin{align}
\frac{dC_{CO_2}}{dt} & = C_{CO_2} \cdot R_{v} + C_{CO_2} \cdot R_{w} - C_{CO_2} \cdot R_o - C_{CO_2} \cdot R_p + C_{CO_2} \cdot R_m \\
\frac{dC_{H_2O}}{dt} & = C_{H_2O} \cdot R_v + C_{H_2O}\cdot R_m + C_{H_2O} \cdot R_e - C_{H_2O}\cdot R_{pr} \\
\frac{dC_{CH_4}}{dt} & = C_{CH_4}\cdot R_w - C_{CH_4}\cdot R_m \\
\frac{dC_{N_2O}}{dt} & = C_{N_2O}\cdot R_w - C_{N_2O} \cdot R_o \\
C_{CO_2}(0) & = C_{CO_2,0}, ~ C_{H_2O}(0) = C_{H_2O,0}, ~ C_{CH_4}(0) = C_{CH_4,0}, ~ C_{N_2O}(0) = C_{N_2O,0}
\end{align}
$$

**Nondimensionalization**

*Following, we observe the dependent variable to be C, and the independent variable to be t. Thus, we convert each to be $C = C^*[C]$ and $t = t^*[t]$. Next, we take the equations defined above, substitute in $C = C^*[C]$ and $t = t^*[t]$, and non-dimensionalize each one by dividing each side of the equations by $\frac{[C]}{[t]}$ to cancel on the [C] on each term. Enabling us to then choose the appropriate scaling factors for [C] and [t] where [t] is the only variable in our equations and [C] can be chosen based on the initial conditions.*

$$
\begin{align}
\frac{[C]}{[t]} \frac{dC_{CO_2}^{*}}{dt^{*}} & = [C]C_{CO_2}^{*} \cdot R_{v} + [C]C_{CO_2}^{*} \cdot R_{w} - [C]C_{CO_2}^{*} \cdot R_{o} - [C]C_{CO_2}^{*} \cdot R_{p} + [C]C_{CO_2}^{*} \cdot R_{m} \\
\frac{[C]}{[t]} \frac{dC_{H_2O}^{*}}{dt^{*}} & = [C]C_{H_2O}^{*} \cdot R_{v} + [C]C_{H_2O}^{*} \cdot R_{m} + [C]C_{H_2O}^{*} \cdot R_{e} - [C]C_{H_2O}^{*} \cdot R_{pr} \\
\frac{[C]}{[t]} \frac{dC_{CH_4}^{*}}{dt^{*}} & = [C]C_{CH_4}^{*} \cdot R_w - [C]C_{CH_4}^{*} \cdot R_m \\
\frac{[C]}{[t]} \frac{dC_{N_2O}^{*}}{dt^{*}} & = [C]C_{N_2O}^{*} \cdot R_{w} - [C]C_{N_2O}^{*} \cdot R_{o} \\
[C]C_{CO_2}^{*}(0) & = C_{CO_2,0}, ~ [C]C_{H_2O}^{*}(0) = C_{H_2O,0}, ~ [C]C_{CH_4}^{*}(0) = C_{CH_4,0}, ~ [C]C_{N_2O}^{*}(0) = C_{N_2O,0}
\end{align}
$$

Then, dividing by $\frac{[C]}{[t]}$ the result is as follows:

$$
\begin{align}
\frac{dC_{CO_2}^{*}}{dt^{*}} & = [t]C_{CO_2}^{*} \cdot R_{v} + [t]C_{CO_2}^{*} \cdot R_{w} - [t]C_{CO_2}^{*} \cdot R_{o} - [t]C_{CO_2}^{*} \cdot R_{p} + [t]C_{CO_2}\cdot R_{m} \\
\frac{dC_{H_2O}^{*}}{dt^{*}} & = [t]C_{H_2O}^{*} \cdot R_{v} + [t]C_{H_2O}^{*} \cdot R_{m} + [t]C_{H_2O}^{*} \cdot R_{e} - [t]C_{H_2O}^{*} \cdot R_{pr} \\
\frac{dC_{CH_4}^{*}}{dt^{*}} & = [t]C_{CH_4}^{*} \cdot R_{w} - [t]C_{CH_4}^{*} \cdot R_{m} \\
\frac{dC_{N_2O}^{*}}{dt^{*}} & = [t]C_{N_2O}^{*} \cdot R_{w} - [t]C_{N_2O}^{*} \cdot R_{o} \\
C_{CO_2}^{*}(0) & = \frac{C_{CO_2,0}}{[C]}, ~ C_{H_2O}^{*}(0) = \frac{C_{H_2O,0}}{[C]}, ~ C_{CH_4}^{*}(0) = \frac{C_{CH_4,0}}{[C]}, ~ C_{N_2O}^{*}(0) = \frac{C_{N_2O,0}}{[C]}
\end{align}
$$

Now, we have to choose our scaling factors.
Let's start by choosing the scaling factor for $[t]$. We choose to let $[t]=\frac{1}{R_{v}}$ as the cancel out the $R_v$ in equations (a) and (b). Note that $R_v$ has the same dimensions as $R_q, R_o, R_p, R_m, R_pr, R_w,$ and $R_e$, so the dimensions on every term will cancel. Then, we choose our scaling factor for [C], since we assume every intial concentration of each gas to be equal we know that $C_{CO_2,0} = C_{H_2O,0} = C_{CH_4,0} = C_{N_2O,0}$, then we can let $[C] = C_{CO_2, 0}$, which will cancel all the intial conditions, making them all equal to one.
This gives us the following equations:

$$
\begin{align}
(a)\quad \frac{dC_{CO_2}^{*}}{dt^{*}} & = C_{CO_2}^{*} + C_{CO_2}^{*} \cdot r_{w} - C_{CO_2}^{*} \cdot r_{o} - C_{CO_2}^{*} \cdot r_{p} + C_{CO_2}^{*}\cdot r_{m} \\
(b)\quad \frac{dC_{H_2O}^{*}}{dt^{*}} & = C_{H_2O}^{*} + C_{H_2O}^{*} \cdot r_{m} + C_{H_2O}^{*} \cdot r_{e} - C_{H_2O}^{*} \cdot r_{pr} \\
(c)\quad \frac{dC_{CH_4}^{*}}{dt^{*}} & = C_{CH_4}^{*} \cdot r_{w} - C_{CH_4}^{*} \cdot r_{m} \\
(d)\quad \frac{dC_{N_2O}^{*}}{dt^{*}} & = C_{N_2O}^{*} \cdot r_{w} - C_{N_2O}^{*} \cdot r_{o} \\
C_{CO_2}^{*}(0) & = 1, ~ C_{H_2O}^{*}(0) = 1, ~ C_{CH_4}^{*}(0) = 1, ~ C_{N_2O}^{*}(0) = 1
\end{align}
$$

Then we let
$$
r_w = \frac{R_w}{R_v}, ~r_o = \frac{R_o}{R_v}, ~r_p = \frac{R_p}{R_v}, ~r_m = \frac{R_m}{R_v}, ~r_e = \frac{R_e}{R_v}, ~r_{pr} = \frac{R_{pr}}{R_v}.
$$

***Visualizing Our Model***

*Now that we have our non-dimensionalized equations we can visualize our model, first importing the necessary packages, defining our variables with the initial condition set to one and with random chosen rates, and then defining the function for each gas. We then create an array of 100 evenly spaced values from zero to three for our t-axis. We then feed this data and each function into the odeint function from the scipy library. We plot the first column output from the odeint function against our t values, namely the mass of each chemical for every value of t.*

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as spi
```

```python
#C02 calculations
C_co2 = 1
r_w = 5
r_o = 2
r_p = 4
r_m = 2
f_co2 = lambda u,t: np.array([u[1], (1 + r_w -r_o - r_p + r_m)*u[0]])
u0_co2 = [1,30]
t = np.linspace(0,3,100)
u_1 = spi.odeint(f_co2,u0_co2,t)

plt.plot(t,u_1[:,0] + C_co2, label='CO2')

#h20 calculations
C_h20 = 1
r_e = 4
r_pr = 2

U0_h20 = [1, 5]
f_h20 = lambda u,t: np.array([u[1], (1 + r_m + r_e - r_pr)*u[0]])
u_2 = spi.odeint(f_h20,U0_h20,t)

plt.plot(t,u_2[:,0] + C_h20, label='H20')

#CH4 calculations
C_ch4 = 1
f_ch4 = lambda u,t:np.array([u[1], (r_w - r_m)*u[0]])
U0_ch4 = [1, 8]
u_3 = spi.odeint(f_ch4, U0_ch4, t)
plt.plot(t, u_3[:,0] + C_ch4, label='CH4')

#N20 calculations
C_n2o = 1
f_n2o = lambda u,t:np.array([u[1], (r_w - r_o)*u[0]])
U0_n2o = [1, 10]
u_4 = spi.odeint(f_n2o, U0_n2o, t)
plt.plot(t, u_4[:,0] + C_n2o, label='N2O')
plt.legend()
plt.title('Concentration of Greenhouse Gas (C*) in the Atmosphere over Time')
plt.xlabel('Time (t*)')
plt.ylabel('Concentration (c*)')
plt.grid(True)
plt.show()
```

***Analyze and Assess the Model***

*Finally, we draw conclusions from our graph, as seen below.*

We can see that for our solution, we do not reach a steady-state solution for all chemicals. If we did, the steady state would most likely be zero.

This is because, except for the decomposition of methane into carbon dioxide and water vapour, the chemicals have no effect on one another.
They are dependent on the rate of emission and absorption by certain biomes and organisms, from volcanoes to trees, etc.

So we have 2 cases: either a chemical is absorbed more than it is emitted, or vice versa. In the first case, the chemical's concentration in the atmosphere would reach 0 eventually. In the second case, the chemical's concentration would grow exponentially, never reaching a steady state, as we can see in the graph. This applies to each chemical independently of the others, and so our solution never reaches a nontrivial steady state solution.

This solution makes sense in the context of the problem, as we essentially have a "multiple lakes" problem like the one we did in class, except the water is continually pumped in. This leads to the continuous growth we see in our solution, as opposed to the steady states we saw in our in-class examples.

This solution is not realistic in the context of our real-world environment, however, as the concentration chemicals in the atmosphere do not reach 0 as time goes to infinity, nor do they grow exponentially. If they did, we would have a much larger problem with global warming than the on we already have.

This suggests that the problem wasn't realistic and required a lot more interactions between chemicals in order to more accurately simulate what actually occurs in the atmosphere, and provide us with a more realistic solution.

In our model, we took liberty of some assumptions - the main one being that all chemicals are released in equal amounts by the various processes. We took this assumption in order to simplify our model significantly, allowing us to use fewer variables and therefore have fewer dimensionless parameters by the end of the nondimensionalization process.

In iterating on this model in the future, we could reassess whether we want to make this assumption, as removing it would result in a more accurate model, accounting for how much of each chemical is added/removed from the atmosphere for each process, and would allow us to change more parameters to get a more varied result.

For this project, however, we felt the number of variables we had was sufficient, and we felt that it wouldn't have a significant effect on the overarching result that we obtained.


The other assumptions we made were all to remove unnecessary external factors. To do this, we made the assumption that the atmosphere was a well-mixed fluid with constant mass and volume, to ensure that the chemicals were treated as if they had no effect on the mass and volume of the atmosphere, and that they dissolved instantly into the atmosphere.
Additionally, we took the assumption that there are no outside factors on the concentration of the chemicals.

Finally, we assumed that the rates of all processes were constant. This would be an interesting topic to develop our model further, as the rates of chemical processes are not constant in real life, and some are in fact very varied, such as the rate of volcanic eruptions. However, for this simpler model, we felt comfortable assuming all rates were constant to ease our calculations, and because we felt that it was outside the scope of this assignment.

Below, you can see some of the models we did to further our understanding of our model. In each of them, we compared a different set of chemicals to understand which ones are growing faster compared to the others in our model.

Each of the graphs is plotted on a "time vs. concentration" scale.

From looking at these graphs, we can see that the rate of growth of concentration of each of the chemicals follows this pattern:

\begin{align}
\text{CH}_{4}>\text{H}_{2}\text{O}>\text{CO}_{2}>\text{N}_{2}\text{O}

This shows us that we might need to further tweak our initial conditions, as in real-life, methane gas is not produced at such high rates. Ideally, it would be below CO$_{2}$ but just above N$_{2}$O.
